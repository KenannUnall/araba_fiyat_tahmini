```python
# Bu kısımda veri setimizin bazı kolonlarında bulunan json veri tipindeki değerlerin ayrıştırılması işlemini gerçekleştiriyoruz.
# Toplam 3 kolon json formatında bilgi taşımaktadır.
import ast
liste=[]
liste2=[]
liste3=[]

for i in df.ilan_kategorisi:
    satir_labels=[]
    i=ast.literal_eval(i)
    for j in i:
        if j["label"]=="A3": # A3 modellerinde ayrıştırma yaparken kayma olmasından dolayı. A3 değerini listeye eklemiyoruz.
            continue
        satir_labels.append(j["label"])
    liste.append(satir_labels)

for i in df.arac_tagleri:
    liste2.append(ast.literal_eval(i))

for i in df.arac_ozellikleri:
    liste3.append(ast.literal_eval(i))
```

```python
liste2[:5] # arac_tagleri kolonundaki json formatındaki değerlerin ayrıştırılmış hali.
```

```python
liste3[:5] #arac_ozellikleri kolonundaki json formatındaki değerlerin ayrıştırılmış hali.
```

```python
# Aşağı kısımda bu listelerin dataframe formatına çevrilip colon isimlerinin belirlenmesi bulunmakta.
df1 = pd.DataFrame(liste, columns=["ilan_kat1","ilan_kat2","ilan_kat3","ilan_kat4","ilan_kat5","ilan_kat6"])
df2 = pd.DataFrame(liste2)
df3 = pd.DataFrame(liste3,columns=["km","yil_model","yakit_turu","motor_gucu(hp)","vites_turu"])

# Dataframelerin birleştirilmesi ve önem taşımayan kolonların kaldırılması işlemi yapılmaktadır.
df_concat = pd.concat([df, df1, df2, df3], axis=1)
df_concat =df_concat.drop(["ilan_basligi","ilan_konumu","ilan_id","fiyat_kuru",
                           "ilan_kategorisi","arac_tagleri","arac_ozellikleri"],axis=1)

# Bu kısımda json formattaki verilerin ayrıştırılması aşamasında kayma olduğundan dolayı tespit edilen değerlerin hepsi
# Renault'a ait olduğundan dolayı Marka değerleri Renault ile doldurulmuştur.
df_concat["Marka"].fillna("Renault",inplace=True)
 
# Silindir hacmi olarak Model sütunundalki sayısal değerler alınmaktadır. Ve Audi markalarına ait farklı tür olarak adlandırılan
# Silindir hacimleri ortak bir değere dönüştürülmüştür.
df_concat['silindir_hacmi'] = df_concat.apply(lambda row: '1.5' if row['Marka'] == 'Audi' and row['Model'].split()[2] == '35'
                                    else ('1.0' if row['Marka'] == 'Audi' and row['Model'].split()[2] == '30' 
                                    else (row['Model'].split()[2] if row['Marka'] == 'Audi' 
                                    else row['Model'][:3])), axis=1)

#ilan_kat1 hepsinde aynı, ilan_kat3 = "marka" sütunu aynı biri kaldırılacak, ilan_kat4 ile "seri" sütunu aynı 
# model sütunu = ilan_kat5 + ilan_kat6 aynı
# İhtiyacımız kalmayan benzeri bulunan kolonları sildik
df_concat =df_concat.drop(["ilan_kat1","ilan_kat3","Seri","ilan_kat5","ilan_kat6","Model"],axis=1)
df_concat=df_concat[df_concat['vites_turu'].notna()]

# ilam_tarihi kolonunu gün-ay-yıl formatına çevirme işlemi
df_concat["ilan_tarihi"]=pd.to_datetime(df_concat['ilan_tarihi']).dt.strftime('%Y-%m-%d')
# km kolonundan sayısal değerin alınıp float türüne çevrilmesi  
df_concat['km'] = df_concat['km'].str.replace(r'\D', '', regex=True).astype("float")
# yil_model kolonundan sayısal değerin alınıp float türüne çevrilmesi
df_concat['yil_model'] = df_concat['yil_model'].str.replace(r'\D', '', regex=True).astype("float")

# 3 adet kaymalardan oluşan silindir_hacmi değerlerine gerçek değerlerinin atanma işlemi.
# ve float veri tipine dönüştürülmesi.
df_concat.loc[df_concat["silindir_hacmi"] == "Exc" , "silindir_hacmi"] = 1.5
df_concat.loc[df_concat["silindir_hacmi"] == "Aut", "silindir_hacmi"] = 1.5
df_concat.loc[df_concat["silindir_hacmi"] == "Joy", "silindir_hacmi"] = 1.3
df_concat["silindir_hacmi"]=df_concat["silindir_hacmi"].astype("float")

# motor_gucu(hp) kolonundan sayısal değerlerin alınması ve float değere dönüştürülmesi.
# Aşağıdaki fonksiyon 255-275 gibi değerlerin '-' işaretinin sağ tarafındaki değeri(max) alır.
# Normal 110 gibi değer ise direkt değeri alır.
def parse_power(value):
    parts = value.split('-')
    if len(parts) > 1:
        parsed_value = int(parts[1].split()[0])
    else:
        parsed_value = int(value.split()[0])
    return parsed_value

# 'parse_power' fonksiyonun 'motor_gucu(hp)' adlı kolona uygulanması ve bu kolonun veri tipinin float'a dönüştürülmesi.
df_concat['motor_gucu(hp)'] = df_concat['motor_gucu(hp)'].apply(parse_power).astype("float")

# Az sayıda bulunan değerlerin veri setinden atılması.
df_concat=df_concat[~(df_concat["ilan_kat4"]=="A5")]
df_concat=df_concat[~(df_concat["yakit_turu"]=="Elektrik")]

# kolonların daha anlamlı başlıklar ile yeniden isimlendirilmesi.
df_concat.rename(columns={'ilan_kat2': 'tasit_turu', 'ilan_kat4': 'model','Marka':'marka'}, inplace=True)
```

## YENİ ÖZNİTELİK OLUŞTURMA
Bu kısımda 3 adet yeni öznitelik oluşturulmuştur. <br> <br>
Beygir Gücü = motor gücü / silindir hacmi <br> <br>
Araç Yaşı = Bulunduğumuz yıl (2023) – yil_model <br> <br>
Yıpranma Oranı = km / (araç yaşı + 0.1) <br> <br>
Burada yıl olarak tahminin yapılacağı yıl alınmaktadır. datatime adlı kütüphane ile şuan da 
bulunduğumuz yılı aldık. Yıpranma oranının hesaplanmasında 0.1 eklenmesinin nedeni paydanın 0 yani 
sıfır bir araç geldiğinde yıpranma oranının değeri sonsuz olmasıdır. 

```python
import datetime
df_concat['beygir_gucu'] = df_concat['motor_gucu(hp)'] / df_concat['silindir_hacmi']
df_concat['arac_yasi'] =  (datetime.datetime.now().year - df_concat['yil_model'])
df_concat['yipranma_orani'] = (df_concat['km']/(df_concat['arac_yasi']+0.1))
```

## SCALE İŞLEMİ
Bu kısımda, X ve Y değeri numpy dizisine çevrilmiş ve ardından scale işlemi gerçekleştirilmiştir. <br>
Scale İşlemi (Ölçeklendirme): Scale işlemi, veri noktalarını belirli bir aralığa dönüştürmek için 
kullanılan bir veri ön işleme tekniğidir. Genellikle veri setinin farklı özelliklerinin farklı ölçeklerde 
olması durumunda kullanılır. Örneğin, bir özellik 0-1 aralığında diğer özellik ise 100-1000 aralığında 
olabilir. Bu durumda scale işlemi kullanılarak her iki özelliği aynı ölçekte temsil etmek mümkündür.<br>
Min-Max Scale Yöntemi: Min-Max scale yöntemi, scale işlemi için yaygın olarak kullanılan bir 
yöntemdir. Bu yöntemde, veri noktaları belirli bir aralığa dönüştürülür. Genellikle 0-1 aralığı tercih 
edilir.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

## TRAIN – TEST SPLIT
Bu kısımda test_size, veri setindeki veri sayısının yüksek olmasından dolayı 0.1 olarak belirlenmiştir. <br>
Train set shape - X_train: (500013, 20) Y_train: (500013, 1) <br>
Test set shape - X_test: (55557, 20) Y_test: (55557, 1)<br>


```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
print("Train set shape - X_train:", X_train.shape, " Y_train:", Y_train.shape)
print("Test set shape - X_test:", X_test.shape, " Y_test:", Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor

# Linear Regression
linear_regression = LinearRegression()
linear_regression.fit(X_train, Y_train)
Y_pred_linear = linear_regression.predict(X_test)
mse_linear = mean_squared_error(Y_test, Y_pred_linear)
r2_linear = r2_score(Y_test, Y_pred_linear)
print('Linear Regression:')
print(f'MSE: {mse_linear}')
print(f'R2 Score: {r2_linear}')
print('')

# Ridge Regression
ridge_regression = Ridge()
ridge_regression.fit(X_train, Y_train)
Y_pred_ridge = ridge_regression.predict(X_test)
mse_ridge = mean_squared_error(Y_test, Y_pred_ridge)
r2_ridge = r2_score(Y_test, Y_pred_ridge)
print('Ridge Regression:')
print(f'MSE: {mse_ridge}')
print(f'R2 Score: {r2_ridge}')
print('')

# Lasso Regression
lasso_regression = Lasso()
lasso_regression.fit(X_train, Y_train)
Y_pred_lasso = lasso_regression.predict(X_test)
mse_lasso = mean_squared_error(Y_test, Y_pred_lasso)
r2_lasso = r2_score(Y_test, Y_pred_lasso)
print('Lasso Regression:')
print(f'MSE: {mse_lasso}')
print(f'R2 Score: {r2_lasso}')
print('')

# Random Forest Regression
random_forest_regression = RandomForestRegressor()
random_forest_regression.fit(X_train, Y_train)
Y_pred_rf = random_forest_regression.predict(X_test)
mse_rf = mean_squared_error(Y_test, Y_pred_rf)
r2_rf = r2_score(Y_test, Y_pred_rf)
print('Random Forest Regression:')
print(f'MSE: {mse_rf}')
print(f'R2 Score: {r2_rf}')
print('')

# K-Nearest Neighbors Regression
knn_regression = KNeighborsRegressor()
knn_regression.fit(X_train, Y_train)
Y_pred_knn = knn_regression.predict(X_test)
mse_knn = mean_squared_error(Y_test, Y_pred_knn)
r2_knn = r2_score(Y_test, Y_pred_knn)
print('K-Nearest Neighbors Regression:')
print(f'MSE: {mse_knn}')
print(f'R2 Score: {r2_knn}')
print('')

# Gradient Boosting Regression
gradient_boosting_regression = GradientBoostingRegressor()
gradient_boosting_regression.fit(X_train, Y_train)
Y_pred_gb = gradient_boosting_regression.predict(X_test)
mse_gb = mean_squared_error(Y_test, Y_pred_gb)
r2_gb = r2_score(Y_test, Y_pred_gb)
print('Gradient Boosting Regression:')
print(f'MSE: {mse_gb}')
print(f'R2 Score: {r2_gb}')
print('')
```
